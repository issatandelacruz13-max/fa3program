```python
def calculate_grade(score):

    if 90 <= score <= 100:
        return 'A'
    elif 80 <= score <= 89:
        return 'B'
    elif 70 <= score <= 79:
        return 'C'
    elif score < 70:
        return 'F'
    else:
        return 'Invalid Score'
```
